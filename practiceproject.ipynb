for some reason there is a column called unnamed:0

```python
dfTrain = pd.read_csv("train.csv")
dfTrain.drop(["Unnamed: 0"],axis=1, inplace=True)
dfTrain
```

```python
dfTest = pd.read_csv("test.csv")
dfTest.drop(["Unnamed: 0"],axis=1, inplace=True)
dfTest
```

```python
def oneHotBind(df, features):
    dummies = pd.get_dummies(df[features])
    res = pd.concat([df, dummies], axis=1)
    res = res.drop(features, axis=1)
    
    return res
```

```python
dfTrain = oneHotBind(dfTrain, ["workclass", "education", "maritalStatus", "occupation", "relationship", "race", "sex", "nativeCountry", "income"])
dfTrain
```

```python
dfTest = oneHotBind(dfTest, ["workclass", "education", "maritalStatus", "occupation", "relationship", "race", "sex", "nativeCountry", "income"])
dfTest
```

```python
for attributes in dfTrain.keys():
    if attributes not in dfTest.keys():
        print("Adding missing feature {}".format(attributes))
        dfTest[attributes] = 0
```

```python
features = []
target = []
for col in dfTrain.columns:
    if col == "income_<=50K" or col == 'income_>50K':
        target.append(col)
    else:
        features.append(col)

print(target)
print(features)
```

```python
x = dfTrain[features]
y = dfTrain[target]
X = dfTest[features]
Y = dfTest[target]
# X_train, X_test, y_train, y_test = train_test_split(x,y)
tree = DecisionTreeClassifier()
tree.fit(x, y)
predictions = tree.predict(X)
print("Accuraccy: " + str(accuracy_score(Y, predictions)))
print(classification_report(Y, predictions))
```

Use Bernoulli Naive Bayes since we are trying to classify if the person makes more or less than 50k a year. We chose Bernoulli since we have one-hot encoded our data, so it is binarized already. 

```python
from sklearn.naive_bayes import BernoulliNB

y = y.iloc[:, 1:]
Y = Y.iloc[:, 1:]
bnb = BernoulliNB()
bnb.fit(x, y)
predictions = bnb.predict(X)

print('Accuracy: '+ str(accuracy_score(Y, predictions)))
print(classification_report(Y, predictions))
```

#SVM

```python
from sklearn import svm

classifier = svm.SVC(gamma='auto')
classifier.fit(x,y)
predictions = classifier.predict(X)
print('Accuracy: '+ str(accuracy_score(Y, predictions)))
print(classification_report(Y, predictions))
```
