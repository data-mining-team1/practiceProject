```python
import numpy as np
import pandas as pd
```

# Importing the dataset

```python
dfTrain = pd.read_csv('adult.data', skipinitialspace=True)
dfTrain.columns = ["age" , "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
dfTest = pd.read_csv('adult.test', skipinitialspace=True)
dfTest.columns = ["age" , "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
dfTest["income"] = dfTest["income"].str.replace('.', '')
```

# Remove records with unknown values

```python
dfTrain = dfTrain[(dfTrain.values != '?').all(axis=1)]
dfTest = dfTest[(dfTest.values != '?').all(axis=1)]
```

```python
print('train length = ', len(dfTrain))
print('test length = ', len(dfTest))
```

# Encoding categorical data

```python
def oneHotBind(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[feature_to_encode])
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop(feature_to_encode, axis=1)
    return (res)
```

```python
dfTrain = oneHotBind(dfTrain, ["workclass", "education", "marital-status", "occupation", "relationship", "race", "sex", "native-country"])
dfTest = oneHotBind(dfTest, ["workclass", "education", "marital-status", "occupation", "relationship", "race", "sex", "native-country"])
```

```python
dfTrain.head()
```

# Transform numerical attribute

```python
def numericalBinary(dataset, features):
    dataset[features] = np.where(dataset[features] >= dataset[features].mean(), 1, 0)
```

```python
numericalBinary(dfTrain, ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week'])
numericalBinary(dfTest, ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week'])
```

```python
dfTrain.head()
```

# Encoding the dependent variable

```python
from sklearn.preprocessing import LabelEncoder

def encode_income(dataset):
    le = LabelEncoder()
    dataset['income'] = le.fit_transform(dataset['income'])
    return dataset

encode_income(dfTrain)
encode_income(dfTest)
```

```python
X_train = dfTrain.loc[:, dfTrain.columns != 'income'].values
y_train = dfTrain['income'].values
X_test = dfTest.loc[:, dfTest.columns != 'income'].values
y_test = dfTest['income'].values
```

# Building k-means clustering algorithm over the training set

## k value = 3

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3)
kmeans.fit(X_train)
```

```python
from sklearn import metrics
score = metrics.accuracy_score(y_train, kmeans.predict(X_train))
print('Accuracy: {0:f}'.format(score))
```

```python
kmeans.cluster_centers_
```

## k value = 5

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(X_train)
```

```python
score = metrics.accuracy_score(y_train, kmeans.predict(X_train))
print('Accuracy: {0:f}'.format(score))
```

```python
kmeans.cluster_centers_
```

## k value = 10

```python
kmeans = KMeans(n_clusters=10)
kmeans.fit(X_train)
```

# Applying the kNN algorithm on the last 10 records from the test set

```python
dfTest = dfTest[-10:]
dfTest
```

```python
for attributes in dfTrain.keys():
    if attributes not in dfTest.keys():
        print('Adding missing feature {}'.format(attributes))
        dfTest[attributes] = 0
        
dfTest = dfTest[dfTrain.keys()]
```

```python
X_test = dfTest.loc[:, dfTest.columns != 'income'].values
y_test = dfTest['income'].values
```

## k value = 3

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)
predictions = knn.predict(X_test)
```

```python
print('Accuracy: ' + str(accuracy_score(y_test, predictions)))
print(classification_report(y_test, predictions))
```
